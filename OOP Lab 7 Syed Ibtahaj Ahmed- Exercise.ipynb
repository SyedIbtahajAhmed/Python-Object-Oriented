1- Mohammad Ali Jinnah University purchases Multiple robots to perform 02 major operations for the students and faculty. For example: 'Cleanobot' performs cleaning of the doorways and corridors of the university. 'Secubot' is an automated robot that stops students from entering the university without their ID cards. Think of more robots that might perform different duties. Using concepts of Composition, create a Generic Robot class with attributes of your own choice. Create another class Department. Create different types of robot instances within this class that use the Robot class attributes. Presnet your output in a readable format where it shows the functionality of a particular robot and to which department it might belong.

```python
class InvalidRobot(Exception):
    pass
class Robot:
    def __init__(self, name = None):
        self.name = name
    def Cleanobot(self):
        return "Performs Cleaning Of The Doorways and Corridors Of The University!"
    def Secubot(self):
        return "is an automated robot that stops students from entering the university without their ID Cards."
class Department:
    def __init__(self, name = None, depart = None):
        self.depart = None
        self.rob = Robot(name)
    def take_inputs(self):
        self.depart = str(input("Enter Department: "))
        self.rob.name = str(input("Enter Robot Name: "))
    def Display(self):
        try:
            if(self.rob.name == "Cleanobot" or self.rob.name == "cleanobot"):
                print("\n\n{} {}".format(self.rob.name, self.rob.Cleanobot()))
                print("\nIt belongs to : {} department.".format(self.depart))
            elif(self.rob.name == "Secubot" or self.rob.name == "secubot"):
                print("\n\n{} {}".format(self.rob.name, self.rob.Secubot()))
                print("\nIt belongs to : {} department.".format(self.depart))
            else:
                raise InvalidRobot
        except(InvalidRobot):
            print("The Robot You Entered Is Not Valid Or Spelled Incorrect.")
        finally:
            print("Thanks For Using!")

anykey = str(input("Press Any Key To Continue or Q To Quit: "))
while(anykey != 'q'):
    obj = Department()
    obj.take_inputs()
    obj.Display()
    anykey = str(input("Press Any Key To Continue or Q To Quit: "))
```
