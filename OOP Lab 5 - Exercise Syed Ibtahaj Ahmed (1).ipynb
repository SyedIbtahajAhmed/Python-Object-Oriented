1- In Football, players have different roles such as Striker, Defender, Mid Fielder and Goalie etc. Performance of each player must be judged according to its position. For instance, the performance of a goalie may be judged on the basis of his saves and not goals scored. Similarly, each player type has special areas where their performance must be judged. Using polymorpishm concepts, create a program where the Performance method can be used as an Abstract method. The program should be able to display performances of each type of player. You can set any attributes for each sub class.

```python
class Players:
    def __init__(self, name = None, perform = None):
        self.name = name
        self.perform = perform
    def take_inputs(self):
        raise NotImplementedError("Subclass Should Implement Performance Method!")
    def Performance(self):
        raise NotImplementedError("Subclass Should Implement Performance Method!")
        
class Striker(Players):
    def __init__(self, goals = None, attempts = None, name = None, perform = None):
        Players.__init__(self, name, perform)
        self.goals = goals
        self.attempts = attempts
    def take_inputs(self):
        self.name = str(input("Enter Name Of The Player: "))
        self.goals = int(input("Enter Goals Scored Of The Striker: "))
        self.attempts = int(input("Enter Total Attempts Of Goals: "))
    def display(self):
        print("\nThe Name Of The Player Is: {}".format(self.name))
        print("The Performance Of {} Is: scored {} out of {}".format(self.name, self.goals, self.attempts))
        print("Success Rate Of {} Is: {} %".format(self.name, self.perform))
    def Performance(self):
        self.perform = self.goals / self.attempts * 100
        return self.perform
class Mid_fielder(Players):
    def __init__(self, assist = None, attempts = None, name = None, perform = None):
        Players.__init__(self, perform = None)
        self.assist = assist
        self.attempts = attempts
    def take_inputs(self):
        self.name = str(input("Enter Name Of The Player: "))
        self.assist = int(input("Successfull assist in which goals were scored: "))
        self.attempts = int(input("Enter Total Attempts Of Shots that were assisted by {} : ".format(self.name)))
    def display(self):
        print("\nThe Name Of The Player Is: {}".format(self.name))
        print("The Performance Of {} Is: Assissted {} out of {}".format(self.name, self.assist, self.attempts))
        print("Success Rate Of {} Is: {} %".format(self.name, self.perform))
    def Performance(self):
        self.perform = self.assist / self.attempts * 100
        return self.perform
    
class Defender(Players):
    def __init__(self, goals = None, attempts = None , name = None, perform = None):
        Players.__init__(self, name, perform)
        self.goals = goals
        self.attempts = attempts
    def take_inputs(self):
        self.name = str(input("Enter Name Of The Player: "))
        self.goals = int(input("Enter Goals defended by The {} : ".format(self.name)))
        self.attempts = int(input("Enter Total Attempts Of Goals from Opposite Team: "))
    def display(self):
        print("\nThe Name Of The Player Is: {}".format(self.name))
        print("The Performance Of {} Is: Defended {} out of {}".format(self.name, self.goals, self.attempts))
        print("Success Rate Of {} Is: {} %".format(self.name, self.perform))
    def Performance(self):
        self.perform = self.goals / self.attempts * 100
        return self.perform
    
class Goal_K(Players):
    def __init__(self, goals = None, attempts = None , name = None, perform = None):
        Players.__init__(self, name, perform)
        self.goals = goals
        self.attempts = attempts
    def take_inputs(self):
        self.name = str(input("Enter Name Of The Player: "))
        self.goals = int(input("Enter Goals Saved by The {} : ".format(self.name)))
        self.attempts = int(input("Enter Total Attempts Of Goals from Opposite Team: "))
    def display(self):
        print("\nThe Name Of The Player Is: {}".format(self.name))
        print("The Performance Of {} Is: Saved {} out of {}".format(self.name, self.goals, self.attempts))
        print("Success Rate Of {} Is: {} %".format(self.name, self.perform))
    def Performance(self):
        self.perform = self.goals / self.attempts * 100
        return self.perform
def inputs(obj):
    return obj.take_inputs()
def perform(obj):
    obj.Performance()
player = str(input("Enter Player Position: "))
while(player != 0):
    if(player == "Striker" or player == "striker"):
        obj_striker = Striker()
        inputs(obj_striker)
        perform(obj_striker)
        obj_striker.display()
        player = str(input("Enter Player Position Again Or 0 To Quit: "))
    elif(player == "midfielder" or player == "Midfielder"):
        obj_fielder = Mid_fielder()
        inputs(obj_fielder)
        perform(obj_fielder)
        obj_fielder.display()
        player = str(input("Enter Player Position Again Or 0 To Quit: "))
    elif(player == "defender" or player == "Defender"):
        obj_def = Defender()
        inputs(obj_def)
        perform(obj_def)
        obj_def.display()
        player = str(input("Enter Player Position Again Or 0 To Quit: "))
    elif(player == "goalkeeper" or player == "Goalkeeper"):
        obj_keeper = Goal_K()
        inputs(obj_keeper)
        perform(obj_keeper)
        obj_keeper.display()
        player = str(input("Enter Player Position Again Or 0 To Quit: "))
```

2- Using Polymorphism, write a program to generate the following shapes:

* Circle
* Triangle
* Square

Create Abstract class and methods and bind the subclasses to use these methods. Create at least three different objects with parameters of your choice.

```python
class Shapes:
    def __init__(self, shape = None, area = None):
        self.shape = shape
        self.area = area
    def take_input(self):
        raise NotImplementedError("Subclass Should Implement Performance Method!")
    def calc_area(self):
        raise NotImplementedError("Subclass Should Implement Performance Method!")
    def Display(self):
        raise NotImplementedError("Subclass Should Implement Performance Method!")

class Circle(Shapes):
    def __init__(self, radius = None, shape = None, area = None):
        Shapes.__init__(self, shape, area)
        self.radius = radius
    def take_input(self):
        self.radius = float(input("Please Enter The Radius Of The Circle: "))
    def calc_area(self):
        self.area = 3.142 * (self.radius * self.radius)
        return self.area
    def Display(self):
        print("\n\nThe Area Of The Circle With Radius {} Is: {}".format(self.radius, self.area))
class Triangle(Shapes):
    def __init__(self, height = None, base = None, shape = None, area = None):
        Shapes.__init__(self, shape, area)
        self.height = height
        self.base = base
    def take_input(self):
        self.height = float(input("Please Enter The Height Of The Triangle: "))
        self.base = float(input("Please Enter The Base Of The Triangle: "))
    def calc_area(self):
        self.area = 1 / 2 * (self.height * self.base)
        return self.area
    def Display(self):
        print("\n\nThe Area Of The Triangle With Height {} And Base {} Is: {}".format(self.height, self.base, self.area))
class Square(Shapes):
    def __init__(self, side = None, shape = None, area = None):
        Shapes.__init__(self, shape, area)
        self.side = side
    def take_input(self):
        self.side = float(input("Please Enter The Length Of The Side Of The Square: "))
    def calc_area(self):
        self.area = self.side * self.side
        return self.area
    def Display(self):
        print("\n\nThe Area Of The Square With Length Of The Side {} Is: {}".format(self.side, self.area))
def inputs(obj):
    return obj.take_input()
def area(obj):
    return obj.calc_area()
def printing(obj):
    return obj.Display()
shape = str(input("Enter The Shape To Calculate Area: "))
while(shape != 0):
    if (shape == "Circle" or shape == "circle"):
        cir = Circle()
        inputs(cir)
        area(cir)
        printing(cir)
        shape = str(input("\nEnter The Shape Again To Calculate Area Or 0 To Quit: "))
    elif (shape == "Triangle" or shape == "triangle"):
        tri = Triangle()
        inputs(tri)
        area(tri)
        printing(tri)
        shape = str(input("\nEnter The Shape Again To Calculate Area Or 0 To Quit: "))
    elif (shape == "Square" or shape == "square"):
        sqr = Square()
        inputs(sqr)
        area(sqr)
        printing(sqr)
        shape = str(input("\nEnter The Shape Again To Calculate Area Or 0 To Quit: "))
```
