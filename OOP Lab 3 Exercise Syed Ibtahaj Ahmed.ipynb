1- A Polygon is a figure having at least three sides. Using principles of inheritance, write a program that can calculate the area of:
* Triangle
* Rectangle
* Square

Include a method to return vertices for each figure.

```python
class Polygon:
    def __init__(self,shape = None, area = None):
        self.area = area
        self.shape = shape
    def vert(self):
        if(self.shape == "Triangle" or self.shape == "triangle"):
            return 3
        elif(self.shape == "Square" or self.shape == "square"):
            return 4
        elif(self.shape == "Rectangle" or self.shape == "rectangle"):
            return 4
        else:
            print("Invalid Polygon!")
            
class Triangle(Polygon):
    def __init__(self,shape, height = None, base = None, area = None):
        Polygon.__init__(self, area = None, shape = None)
        self.height = height
        self.base = base
    def area_tri(self):
        self.area = 1 / 2 * base * height
        return self.area
    
class Square(Polygon):
    def __init__(self,shape, length = None, area = None):
        Polygon.__init__(self, area = None, shape = None)
        self.length = length
    def area_sq(self):
        self.area = self.length * self.length
        return self.area
    
class Rectangle(Polygon):
    def __init__(self,shape, length = None, width = None, area = None):
        Polygon.__init__(self, area = None, shape = None)
        self.length = length
        self.width = width
    def area_re(self):
        self.area = self.length * self.width
        return self.area
########################################################################        
shape = str(input("Enter Shape: "))
while(shape != 'q'):
    if(shape == "Triangle" or shape == "triangle"):
        height = float(input("Enter Height: "))
        base = float(input("Enter Base: "))
        t1 = Triangle(shape, height, base)
        print("Area Of Trianlge Is : ", t1.area_tri())
        t1_vert = Polygon(shape)
        print("The Vertices Of The Triangle Is: ", t1_vert.vert())
        shape = str(input("Enter Shape Again or Any Other Key To Quit: "))
        
    elif(shape == "Square" or shape == "square"):
        length = float(input("Enter Length Of The Square: "))
        sq1 = Square(shape, length)
        print("Area Of Square Is: ", sq1.area_sq())
        sq1_vert = Polygon(shape)
        print("The Vertices Of The Square Is: ", sq1_vert.vert())
        shape = str(input("Enter Shape Again or Any Other Key To Quit: "))
        
    elif(shape == "Rectangle" or shape == "rectangle"):
        length = float(input("Enter Length Of The Rectangle: "))
        width = float(input("Enter Width Of The Rectangle: "))
        re1 = Rectangle(shape, length, width)
        print("Area Of Square Is: ", re1.area_re())
        re1_vert = Polygon(shape)
        print("The Vertices Of The Rectangle Is: ", re1_vert.vert())
        shape = str(input("Enter Shape Again or Any Other Key To Quit: "))
    else:
        print("Invalid Shape!")
        shape = str(input("Enter Shape Again or Any Other Key To Quit: "))

print("Thanks For Using")
```

2- Resistors, capacitors and Inductors may belong to a superclass because of some similar attributes such as Impedance. Create a super class for these components. Create methods to return impedance of the respective component. Assume frequency of input to be 50 Hertz. You can use math library to include necessary constants.
* Inductor Impedance = 2.pi.f.L
* Capacitor Impedance = 1/(2.pi.f.C)

```python
class Components:
    def __init__(self,tol = None, tol1 = None, tol2 = None, imp = None):
        self.imp = imp
        self.tol = tol
        self.tol1 = tol1
        self.tol2 = tol2
    
class Capacitor(Components):
    def __init__(self, capacitance, tol = None, tol1 = None, tol2 = None, imp = None):
        Components.__init__(self,tol, tol1, tol2, imp)
        self.capacitance = capacitance
    def calc_cap(self):
        self.imp = 1 / 2 * 3.142 * 50 * self.capacitance
        return self.imp
    def tolerance(self):
        self.tol = (self.tol / 100) * self.imp
        self.tol1 = self.imp + self.tol
        self.tol2 = self.imp - self.tol
        return self.tol1, self.tol2
    
class Inductor(Components):
    def __init__(self, inductance, tol = None, tol1 = None, tol2 = None, imp = None):
        Components.__init__(self, tol, tol1, tol2, imp)
        self.inductance = inductance
    def calc_ind(self):
        self.imp = 2 * 3.142 * 50 * self.inductance
        return self.imp
    def tolerance(self):
        self.tol = (self.tol / 100) * self.imp
        self.tol1 = self.imp + self.tol
        self.tol2 = self.imp - self.tol
        return self.tol1, self.tol2

class Resistor(Components):
    def __init__(self, resistance, tol = None, tol1 = None, tol2 = None, imp = None):
        Components.__init__(self,tol, tol1, tol2, imp)
        self.resistance = resistance
    def print_res(self):
        return self.resistance
    def tolerance(self):
        self.tol = (self.tol / 100) * self.resistance
        self.tol1 = self.resistance + self.tol
        self.tol2 = self.resistance - self.tol
        return self.tol1, self.tol2

comp = str(input("Enter Component To Calculate Impedence: "))
while(comp != 0):
    if(comp == "Capacitor" or comp == "capacitor"):
        capacitance = float(input("Enter Capcitance: "))
        tolerance = float(input("Enter Tolerance: "))
        cap = Capacitor(capacitance, tolerance)
        print("\nThe Impedence Of The Capacitor Is: {}".format(cap.calc_cap()))
        print("\n\nThe Impedence With Tolerence: {}".format(cap.tolerance()))
        comp = str(input("Enter Component To Calculate Impedence: "))
    elif(comp == "Inductor" or comp == "inductor"):
        inductance = float(input("Enter Inductance: "))
        tolerance = float(input("Enter Tolerance: "))
        ind = Inductor(inductance, tolerance)
        print("\nThe Impedence Of The Inductor Is: {}".format(ind.calc_ind()))
        print("\n\nThe Impedence With Tolerence: {}".format(ind.tolerance()))
        comp = str(input("Enter Component To Calculate Impedence: "))
    elif(comp == "resistor" or comp == "Resistor"):
        resistance = float(input("Enter Resistance: "))
        tolerance = float(input("Enter Tolerance: "))
        res = Resistor(resistance, tolerance)
        print("The Resistance Of The Resistor Is: {}".format(res.print_res()))
        print("The Resistance With Tolerance: {}".format(res.tolerance()))
    else:
        print("Invalid Input!")
        comp = str(input("Enter Component Again To Calculate Impedence: "))
print("Thank You For Using!")
```
